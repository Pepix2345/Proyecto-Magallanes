Librerias a utilizar

```python
import pandas as pd
import itertools 
```

Definiendo las reglas

```python
strategies=['Rock','Paper','Scissors','Lizard','Spock']
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0),
        ('Scissors','Lizard'):(1,0),
        ('Lizard','Scissors'):(0,1),
        ('Lizard','Paper'):(1,0),
        ('Paper','Lizard'):(0,1),
        ('Paper','Spock'):(1,0),
        ('Rock','Lizard'):(1,0),
        ('Lizard','Rock'):(0,1),
        ('Spock','Paper'):(0,1),
        ('Lizard','Spock'):(1,0),
        ('Spock','Lizard'):(0,1),
        ('Spock','Scissors'):(1,0),
        ('Scissors','Spock'):(0,1),
        ('Spock','Rock'):(1,0),
        ('Rock','Spock'):(0,1),
        ('Spock','Spock'):(0,0),
        ('Lizard','Lizard'):(0,0)}
```

Definiendo el objeto 

```python
class Player: 
    def __init__(self, name,score=0, strategy=None):
        self.name=name
        self.score=score
        self.strategy=strategy
    def increase_score(self,value):
        self.score+=value
    def get_strategy(self):
        from random import choice
        self.strategy=choice(strategies)
        return self.strategy
```

Definiendo a los jugadores

```python
Charles=Player("Charles")
Edgard= Player("Edgard")
Rosite=Player("Rosite")
Joss=Player("Joss")
Magallanes=Player("Magallanes")
Burkli=Player("Burkli")
```

```python
nombres= [Charles, Edgard, Rosite, Joss,Magallanes,Burkli]
```

Realizando las interacciones

```python
for i in range(200):
    for player1,player2 in itertools.combinations(nombres,2):
        player1_payoff,player2_payoff=payoff[player1.get_strategy(),player2.get_strategy()]
        player1.increase_score(player1_payoff)
        player2.increase_score(player2_payoff)
    print("ronda:",i+1)
    print(player1.strategy,player2.strategy)
```

Creando la sociedad

```python
vecindad_del_chavo=[name.__dict__ for name in nombres]
```

Creando base de datos

```python
vecindad_df=pd.DataFrame(vecindad_del_chavo)
vecindad_df
```

```python
winner=vecindad_df.score.max()
vecindad_df[vecindad_df.score==winner]
```
