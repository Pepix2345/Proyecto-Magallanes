### Librerias a utilizar

```python
import pandas as pd 
from random import choice
import itertools
```

### Estableciendo las regas

```python
#Estrategias
strategies=['Rock','Paper','Scissors','Lizard','Spock']
#Reglas
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0),
        ('Scissors','Lizard'):(1,0),
        ('Lizard','Scissors'):(0,1),
        ('Lizard','Paper'):(1,0),
        ('Paper','Lizard'):(0,1),
        ('Paper','Spock'):(1,0),
        ('Rock','Lizard'):(1,0),
        ('Lizard','Rock'):(0,1),
        ('Spock','Paper'):(0,1),
        ('Lizard','Spock'):(1,0),
        ('Spock','Lizard'):(0,1),
        ('Spock','Scissors'):(1,0),
        ('Scissors','Spock'):(0,1),
        ('Spock','Rock'):(1,0),
        ('Rock','Spock'):(0,1),
        ('Spock','Spock'):(0,0),
        ('Lizard','Lizard'):(0,0)}
```

### Estableciendo la sociedad

```python
#Nombres de los jugadores
names=['Charles','Edgard','Joseline','Jose', 'Magallanes','Elena','Rosite','Sofi','Politai','Burkli']
```

```python
#Sociedad
society=[{'name':n,'score':0,'strategy':None} for n in names]
# several rounds
for aRound in range(200):

    # en each round:
    for player1,player2 in itertools.combinations(society,2):
        # each chooses strategy
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)

        print(player1['name']+ " chose " +player1['strategy'] +" - "+player2['name']+ " chose "+ player2['strategy'])
        if payoff[player1['strategy'],player2['strategy']]==(1,0):
                print(player1['name']+" won ")
        if payoff[player1['strategy'],player2['strategy']]==(0,1):
                print(player2['name']+ " won ")
        if payoff[player1['strategy'],player2['strategy']]==[0,0]:
                print("empate")
        #resultado 
        result=payoff[player1['strategy'],player2['strategy']]
        #actualiza resultado
        player1['score']+=result[0]
        player2['score']+=result[1]
```

```python
resultados=pd.DataFrame(society)
resultados
```
