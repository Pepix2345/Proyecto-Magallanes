### Librerias a utilizar

```python
import pandas as pd 
from random import choice
import itertools
```

### Estableciendo las reglas

```python
#Estrategias
strategies=['Rock','Paper','Scissors']
#Reglas
payoff={('Rock','Paper'):(0,1),
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
#Decision 
choice(strategies);
```

### Definiendo a los jugadores

```python
#Nombre de los jugadores 
names=['Charles','Joseline','Rosite','Elena']
```

### Estableciendo la sociedad

```python
les_evitatives=[{'name':n,'score':0,'strategy':None} for n in names]

# several rounds
for aRound in range(200):

    # en each round:
    for player1,player2 in itertools.combinations(les_evitatives,2):
        # each chooses strategy
        player1['strategy']=choice(strategies)

        player2['strategy']=choice(strategies)
        print(player1['name'] + " chose " + player1['strategy'] +" - "+player2['name'] + " chose " + player2['strategy'] )
        if payoff[player1['strategy'],player2['strategy']] == (1,0):
                print(player1['name'] + " won")
        if payoff[player1['strategy'],player2['strategy']] == (0,1):
                print(player2['name'] + " won")
        if payoff[player1['strategy'],player2['strategy']]==(0,0):
                print("TIE!")
        #resultado 
        result=payoff[player1['strategy'],player2['strategy']]
        #actualiza resultado
        player1['score']+=result[0]
        player2['score']+=result[1]
```

```python
resultados=pd.DataFrame(les_evitatives)
resultados
```

```python
winner=resultados.score.max()
resultados[resultados.score==winner]
```
